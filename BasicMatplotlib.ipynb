```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.image import imread

print(np.__version__)
```

```python
x = np.arange(0,6,0.1)    #以0.1为单位，生成0到6的数据
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle = '--', label='cos')
plt.xlabel('x')    #x轴标签
plt.ylabel('y')    #y轴标签
plt.title('sin & cos')
plt.legend()
plt.show()
```
